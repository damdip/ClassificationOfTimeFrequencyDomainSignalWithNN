```python
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
sys.path.append('../utils')
from getIndexesFromNumber import getIndexesFromNumber

def getSignalFromFile(file_path):
    """Legge un segnale da un file di testo."""
    signal = np.loadtxt(file_path)
    return signal
```
