# Modifica da segnale nel tempo a segnale nel tempo frequenza

```python
from scipy import signal as sig

# Funzione per convertire un segnale in spettrogramma (matrice 2D)
def signal_to_spectrogram(signal, fs=1024, nperseg=128, noverlap=120):
    """
    Converte un segnale 1D in uno spettrogramma (matrice 2D)
    
    Args:
        signal: array 1D del segnale temporale
        fs: frequenza di campionamento
        nperseg: lunghezza della finestra
        noverlap: overlap tra finestre
    
    Returns:
        Sxx_db: matrice 2D dello spettrogramma in dB
    """
    f, t, Sxx = sig.spectrogram(signal, fs=fs, nperseg=nperseg, noverlap=noverlap)
    Sxx_db = 10 * np.log10(Sxx + 1e-10)  # Converti in dB, evita log(0)
    return Sxx_db
```

# Creazione vettori spettrogrammi di mel

```python
def signal_to_mel_spectrogram(signal):
    """
    Converte un segnale temporale in mel-spettrogramma
    
    Args:
        signal: array 1D del segnale nel tempo
        sr: sample rate (frequenza di campionamento)
        n_fft: lunghezza della FFT window
        hop_length: numero di campioni tra frame consecutivi
        n_mels: numero di bande Mel
        fmin: frequenza minima
        fmax: frequenza massima (None = sr/2)
        
    Returns:
        mel_spectrogram: spettrogramma in scala Mel

    """

    mel_spec = librosa.feature.melspectrogram(
        y=signal,
        sr=48000,
        n_fft=2048,        
        hop_length=64,   
        n_mels=10,       
    )  
    return mel_spec
```

```python

for i in range(1, 1601):
    indexes = getIndexesFromNumber(i)  
    file_path = '../dataset/test_object_2_3mm/' + indexes.split()[0] + '/' + indexes + '.txt'
    signal = getSignalFromFile(file_path)
    mel_spec = signal_to_mel_spectrogram(signal)
    output_dir = '../dataset/test_object_2_3mm_mel_spectrograms/' + indexes.split()[0] + '/'
    os.makedirs(output_dir, exist_ok=True)
    np.save(output_dir + indexes + '.npy', mel_spec)
    print(f"Saved mel-spectrogram for {indexes} at {output_dir + indexes + '.npy'}")
```
