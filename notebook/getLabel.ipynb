```python
# Leggi il file Excel
df = pd.read_excel(excell_path, header=None, sheet_name='3 mm')
first_col = df.iloc[:, 0].dropna()    # Estrai la prima colonna
for col in df.columns[1:]:         # Accoda i valori delle altre colonne (saltando la prima)
    first_col = pd.concat([first_col, df[col].dropna()], ignore_index=False)
    #print(first_col.shape)


label = first_col.to_numpy()
label.shape
```

```python
label[label == 2] = 1
print(label.dtype)  # Output: [0 1 3 1 4]

np.savetxt('labels.txt', label, fmt='%d')  

```
